```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2020,1,1)
end = datetime.datetime(2020,12,31)
```

# Gather Stock Information

```python
#STOCK INFORMATION ON SELECT FIRMS IN THE CONSUMER CYCLICAL INDUSTRY BASED ON MARKET CAP

# JPMorgan Chase & Co.
JPM = data.DataReader('JPM','yahoo',start,end)
# Bank of America Corporation
BAC = data.DataReader('BAC','yahoo',start,end)
# Royal Bank of Canada
RY.TO = data.DataReader('RY.TO','yahoo', start,end)
# Wells Fargo & Company
WFC = data.DataReader('WFC', 'yahoo', start,end)
# Morgan Stanley
MS = data.DataReader('MS','yahoo',start,end)
# Citigroup Inc.
C = data.DataReader('C','yahoo', start,end)
# The Toronto-Dominion Bank
TD.TO = data.DataReader('TD.TO','yahoo', start,end)
# Royal Bank of Canada
RY = data.DataReader('RY','yahoo', start,end)
# HSBC Holdings plc
HSBC = data.DataReader('HSBC','yahoo', start,end)
# Broadcom 
GS = data.DataReader('GS','yahoo', start,end)
```

```python
tickers = ['JPM','BAC','RY.TO','WFC','MS','C', 'TD.TO','RY', 'HSBC' , 'GS'] 
```

```python
financial_stocks = pd.concat([JPM,BAC,RY.TO,WFC,MS,C,TD.TO,RY,HSBC,GS], axis = 1, keys = tickers)
```

```python
financial_stocks.columns.names = ['Financial Services','Stock Info']
```

```python
financial_stocks.head()
```

```python
financial_stocks.to_csv('Financial_Servicesfincial.csv')
```
