```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2020,6,1)
end = datetime.datetime(2020,12,31)
```

# Gather Stock Information

```python
#STOCK INFORMATION ON SELECT FIRMS IN THE CONSUMER CYCLICAL INDUSTRY BASED ON MARKET CAP

# Apple
AAPL = data.DataReader('AAPL','yahoo',start,end)
# Microsoft
MSFT = data.DataReader('MSFT','yahoo',start,end)
# Taiwan Semiconductor 
TSM = data.DataReader('TSM','yahoo', start,end)
# NVIDA
NVDA = data.DataReader('NVDA', 'yahoo', start,end)
# Adobe
ADBE = data.DataReader('ADBE','yahoo',start,end)
# ASML Holding
ASML = data.DataReader('ASML','yahoo', start,end)
# INTEL
INTC = data.DataReader('INTC','yahoo', start,end)
# CRM
CRM = data.DataReader('CRM','yahoo', start,end)
# Cisco
CSCO = data.DataReader('CSCO','yahoo', start,end)
# Broadcom 
AVGO = data.DataReader('AVGO','yahoo', start,end)
```

```python
tickers = ['AAPL','MSFT','TSM','NVDA','ADBE','ASML', 'INTC','CRM' , 'CSCO' , 'AVGO'] 
```

```python
consumer_stocks = pd.concat([AAPL,MSFT,TSM,NVDA,ADBE,ASML,INTC,CRM,CSCO,AVGO], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Technology Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Technology_Service_Sector.csv')
```

# Exploratory Data Analysis
